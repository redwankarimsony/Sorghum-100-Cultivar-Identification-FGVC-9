```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_name = "densenet121"
version = 0
log_dir = os.path.join("logs", model_name, f"version_{version}")
```

```python

metrics = pd.read_csv(f'{log_dir}/metrics.csv')
plt.rcParams['font.size']=18

train_acc = metrics['train_acc'].dropna().reset_index(drop=True)
valid_acc = metrics['valid_acc'].dropna().reset_index(drop=True)

fig = plt.figure(figsize=(7, 6))
plt.grid(True)
plt.plot(train_acc, color="r", marker="o", label='train/acc')
plt.plot(valid_acc, color="b", marker="x", label='valid/acc')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.savefig(f'{log_dir}/acc.png')

train_loss = metrics['train_loss'].dropna().reset_index(drop=True)
valid_loss = metrics['valid_loss'].dropna().reset_index(drop=True)

fig = plt.figure(figsize=(7, 6))
plt.grid(True)
plt.plot(train_loss, color="r", marker="o", label='train/loss')
plt.plot(valid_loss, color="b", marker="x", label='valid/loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.savefig(f'{log_dir}/loss.png')

lr = metrics['lr'].dropna().reset_index(drop=True)

fig = plt.figure(figsize=(7, 6))
plt.grid(True)
plt.plot(lr, color="g", marker="o", label='learning rate')
plt.title("DenseNet-121 LR")
plt.ylabel('LR')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.savefig(f'{log_dir}/lr.png')
```

```python
plt.rcParams['font.size']=18
plt.figure(figsize=(12,10))
for i in range(2):
    model_name = "densenet121"
    version = i
    log_dir = os.path.join("logs", model_name, f"version_{version}")

    metrics = pd.read_csv(f'{log_dir}/metrics.csv')
    

    train_acc = metrics['train_acc'].dropna().reset_index(drop=True)
    valid_acc = metrics['valid_acc'].dropna().reset_index(drop=True)

    # fig = plt.figure(figsize=(7, 6))
    plt.grid(True)
    plt.plot(train_acc, color="r",  label=f'train/acc F-{i}')
    plt.plot(valid_acc, color="b", label=f'valid/acc F-{i}')
plt.title("DenseNet-121 5-Fold Accuracies")
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.savefig(f'{os.path.join("logs", model_name)}/acc-combined.png')
```
